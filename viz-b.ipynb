```python
gammas = sns.load_dataset('gammas')
gammas.head(5)
```

```python
sns.relplot(x='subject', y='BOLD signal', data=gammas)
```

```python
grouping =gammas.groupby('timepoint', as_index=False).sum()
sns.lmplot(x='timepoint', y='BOLD signal', data=grouping)
```

```python
sns.barplot(x='timepoint', y='BOLD signal', data=grouping ,ci=None)
```
